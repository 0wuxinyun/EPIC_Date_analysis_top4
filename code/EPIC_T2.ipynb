# Lemmatizing, preprocessing words

```python
def lemmatize_stemming(text):
    stemmer = WordNetLemmatizer()
    return stemmer.lemmatize(text, pos='v')
def preprocess(text):
    result = []
    for token in gensim.utils.simple_preprocess(text):
        if token not in gensim.parsing.preprocessing.STOPWORDS and len(token) > 3:
            result.append(lemmatize_stemming(token))
    return result
```

# Model based on contents of articles

```python
processed_docs = document['contents'].map(preprocess)
```

```python
processed_docs
```

# Creating dictionary(bag of words) from preprocess

```python
dictionary = gensim.corpora.Dictionary(processed_docs)
count = 0
for k, v in dictionary.iteritems():
    print(k, v)
    count += 1
    if count > 10:
        break
dictionary.filter_extremes(no_below=15, no_above=0.5, keep_n=100000)
```

```python
bow_corpus = [dictionary.doc2bow(doc) for doc in processed_docs]
```

```python
from gensim import corpora, models
tfidf = models.TfidfModel(bow_corpus)
corpus_tfidf = tfidf[bow_corpus]
from pprint import pprint
for doc in corpus_tfidf:
    pprint(doc)
    break
```

# Output of contents

```python
lda_model_tfidf = gensim.models.LdaMulticore(corpus_tfidf, num_topics=5, id2word=dictionary, passes=2, workers=5)
for idx, topic in lda_model_tfidf.print_topics(-1):
    print('Topic: {} Word: {}'.format(idx, topic))
    print()
```

```python
for i,topic in lda_model_tfidf.show_topics(formatted=True, num_topics=5, num_words=30):
    print(str(i)+": "+ topic)
    print()
```

```python
topics = [lda_model_tfidf[corpus_tfidf[i]] for i in range(len(articles))]
```

```python
def topics_document_to_dataframe(topics_document, num_topics):
    res = pd.DataFrame(columns=range(num_topics))
    for topic_weight in topics_document:
        res.loc[0, topic_weight[0]] = topic_weight[1]
    return res
```

```python
document_topic = \
pd.concat([topics_document_to_dataframe(topics_document, num_topics=5) for topics_document in topics]) \
  .reset_index(drop=True).fillna(0)
```
